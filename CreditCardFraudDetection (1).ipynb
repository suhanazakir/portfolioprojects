It is important that credit card companies are able to recognize fraudulent credit card transactions so that customers are not charged for items that they did not purchase.




**Machine Learning and AI** : Modern fraud detection systems often use machine learning algorithms to analyze transaction data and identify patterns that might indicate fraudulent activity. These systems can learn from historical data and adapt to new fraud tactics.




The goal of credit card fraud detection is to protect cardholders and financial institutions from losses due to unauthorized transactions while minimizing disruptions to legitimate cardholders' activities.

```python
#data partitioning
df.head()

```

```python
x = df.drop(columns='Class')
y = df['Class']
```

```python
#since data is highly imbalanced
from imblearn.under_sampling import RandomUnderSampler
undersampler=RandomUnderSampler()
x_under,y_under=undersampler.fit_resample(x,y)
y_under.value_counts()
```

Now the dataframe correctly balanced.

```python
#data spltting
x_train, x_test, y_train, y_test = train_test_split(x_under, y_under, test_size=0.3, random_state=42)
```

```python
#data scaling
minmax = MinMaxScaler()
x_train_scaled = minmax.fit_transform(x_train)
x_test_scaled = minmax.transform(x_test)

```

```python
#modeltraining
models  = [
     {
          'name': 'KNN',
          'model': KNeighborsClassifier()


      },
      {
          'name': 'SVM',
          'model':SVC()


      },
      {
          'name': 'Naive Bayes',
          'model':GaussianNB()

      }


]
```

```python
for i in models:
  i['model'].fit(x_train_scaled,y_train)
  y_pred = i['model'].predict(x_test_scaled)
  print(f"{i['name']}\n")
  print(classification_report(y_test,y_pred))
  print('\n\n')
```

Here,model using knn and svm gave 93% accuracy.so we are going to select knn and do hyperparameter tuning for best results

```python
#cross validation

knn=KNeighborsClassifier()
x_under_scaled = minmax.transform(x_under)
cross_val_score(knn,x_under_scaled,y_under,cv=5,scoring='accuracy')
cross_val_score(knn,x_under_scaled,y_under,cv=5).mean()
```

hyperparameter tuning

```python
knn_model=KNeighborsClassifier()
parameters={'n_neighbors':[i for i in range(1,32,2)],'weights':['uniform','distance']}

randomized_cv=RandomizedSearchCV(knn_model,parameters,cv=5,n_iter=4)
randomized_cv.fit(x_under_scaled,y_under)
```

```python
randomized_cv.best_params_
```

```python
best=randomized_cv.best_estimator_
```

```python
y_pred_best=best.predict(x_test_scaled)
```

```python

print(classification_report(y_test,y_pred_best))
```
